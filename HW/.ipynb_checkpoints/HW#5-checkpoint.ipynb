```python
# Homework:
# Write your own Runge-Kutta or any scheme to calculate the orbit of two equal mass binary.
# The mass of each binary makes GM_{i}=1, (i is 1 or 2). The two binary stars are
# seperated by a distance of 1 initially. Initialize the binary velocity so that
# each binary star will have an eccentricity of 0.0. Evolve this system to 100
# binary orbits, and plot their orbits in the x-y plane. 
```

All we need is the initial velocity. The orbits are circular, $M_{Tot}=2$, and $r=\frac{1}{2}$ so: $v_{circ}=\sqrt{\frac{GM_{Tot}}{r}} \Rightarrow \boxed{v=2}$

However, I think I'm messing up the units of G somehow, thus, $v_{circ}$ should be $\boxed{v=\frac{1}{\sqrt{2}}}$. The period is $vT=2\pi r \Rightarrow \boxed{T=\sqrt{2}\pi}$

Below, I calculate and plot half an orbit so that the plot visually confirms the circular nature and that only half an orbit was covered.

```python
import numpy as np
import matplotlib.pyplot as plt
import time

G=1.
M=1.
t_f=0.5*np.sqrt(2)*np.pi
h=0.001
r1=[0.5,0.,0.,1./np.sqrt(2)]       #[x-coord, Vx, y-coord, Vy]
r2=[-0.5,0.,0.,-1./np.sqrt(2)]
t=0


def f(t, ri, rj):
    f=np.zeros(4)
    d=np.sqrt((ri[0]-rj[0])*(ri[0]-rj[0])+(ri[2]-rj[2])*(ri[2]-rj[2]))
    f[0]=ri[1]
    f[1]=-G*M/d/d/d*(ri[0]-rj[0])
    f[2]=ri[3]
    f[3]=-G*M/d/d/d*(ri[2]-rj[2])
    return f

def rk4(t,h,ri,rj):
    k1 = np.zeros(4)
    k2 = np.zeros(4)
    k3 = np.zeros(4)
    k4 = np.zeros(4)
    k1=h*f(t,ri,rj)
    k2=h*f(t+0.5*h,ri+0.5*k1,rj)
    k3=h*f(t+0.5*h,ri+0.5*k2,rj)
    k4=h*f(t+h,ri+k3,rj)
    ri+=(k1+2*k2+2*k3+k4)/6.
    return ri

xp1=[]
yp1=[]
xp2=[]
yp2=[]

start_time = time.time()    #Examine run time

while(t<t_f):
    if((t+h)>t_f):
        h=t_f-t
    r1 = rk4(t,h,r1,r2)
    r2 = rk4(t,h,r2,r1)
    xp1=np.append(xp1,r1[0])
    yp1=np.append(yp1,r1[2])
    xp2=np.append(xp2,r2[0])
    yp2=np.append(yp2,r2[2])
    t = t+h
print("--- %s seconds ---" % (time.time() - start_time))

plt.plot(xp1,yp1)
plt.plot(xp2,yp2)
plt.axis('scaled')
plt.title('1/2 Orbit')
plt.show()
    
```

```python
G=1.
M=1.
t_f=100.*np.sqrt(2)*np.pi
h=0.01
r1=[0.5,0.,0.,1./np.sqrt(2)]       #[x-coord, Vx, y-coord, Vy]
r2=[-0.5,0.,0.,-1./np.sqrt(2)]
t=0

xp1=[]
yp1=[]
xp2=[]
yp2=[]


start_time = time.time()    #Examine run time
while(t<t_f):
    if((t+h)>t_f):
        h=t_f-t
    r1 = rk4(t,h,r1,r2)
    r2 = rk4(t,h,r2,r1)
    xp1=np.append(xp1,r1[0])
    yp1=np.append(yp1,r1[2])
    xp2=np.append(xp2,r2[0])
    yp2=np.append(yp2,r2[2])
    t = t+h
print("--- %s seconds ---" % (time.time() - start_time))
    
plt.plot(xp1,yp1)
plt.axis('scaled')
plt.title('100 Orbits Mass1')
plt.show()

plt.plot(xp2,yp2)
plt.axis('scaled')
plt.title('100 Orbits Mass2')
plt.show()
```

You can see from the 100 orbit plots above that the binary remains approximately circular but it has a slight drift in its center-of-mass velocity. Note, however, that I shrank the 'h-value' (i.e. the timestep) by a factor of 10 because the simulation took too long.
