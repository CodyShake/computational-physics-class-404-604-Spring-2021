##### Write a program to numerically evaluate the derivative of f(x)=cos x at an arbitrary position x.

###### 1) Evaluate the derivative at position x using the forward difference scheme. Plot log(Error)=log10|f’(numerically)-f’(analytic)| against log10(dx) for dx=1e-16,1e-15,….,0.1 at x=π/4. You should find Error~$dx^n$; what is n? what is the optimal value for dx.

```python
import numpy as np
import matplotlib.pyplot as plt

def dfw(x,dx):
    return (np.cos(x+dx)-np.cos(x))/dx

x=np.pi/4.0
xarr=[]
yarr1=[]

for dx in 10.0**np.arange(-16.,-1.,1):
    xarr.append(dx)
    yarr1.append(np.abs(dfw(x,dx)+np.sin(x)))

plt.plot(xarr,yarr1)
plt.xlabel('dx')
plt.ylabel('Error')
plt.xscale('log')
plt.yscale('log')

plt.show()
```

After $dx$ reaches its optimal value, $Error \approx dx$ so $n=1$. The optimal value is $dx \approx 10^{-8}$.

##### 2) Repeat the exercise using the central difference scheme. Why this scheme is better?

```python
def dce(x,dx):
    return (np.cos(x+dx/2.0)-np.cos(x-dx/2.0))/dx

yarr2=[]

for dx in 10.0**np.arange(-16.,-1.,1):
    yarr2.append(np.abs(dce(x,dx)+np.sin(x)))

plt.plot(xarr,yarr1)
plt.plot(xarr,yarr2)
plt.xlabel('dx')
plt.ylabel('Error')
plt.xscale('log')
plt.yscale('log')

plt.show()
```

The central difference scheme (orange) is better because its error is proportional to $dx^2$ while the forward difference scheme is proportional to $dx$.

##### 3) Repeat (2) for x=0. Why do the results differ? (hint: write out the error term derived in the class explicitly by plugging the derivative of cos or sin)

```python
x=0.0

yarr1=[]
yarr2=[]

for dx in 10.0**np.arange(-16.,-1.,1):
    yarr1.append(np.abs(dfw(x,dx)+np.sin(x)))
    yarr2.append(np.abs(dce(x,dx)+np.sin(x)))

plt.plot(xarr,yarr1)
plt.plot(xarr,yarr2)
plt.xlabel('dx')
plt.ylabel('Error')
plt.xscale('log')
plt.yscale('log')

plt.show()
```

The central difference (orange) is not shown because Error=0 cannot be graphed on a log-plot. In this instance, central difference is EXACT because cosine is an even function where $cos(x)=cos(-x)$. Thus, when we take the difference in the calculation of the derivative, we are given zero, which is the exact value of sine at x=0.

In other words,

$\frac{d}{dx}cos(x)|_{x=0}=\frac{cos(\frac{h}{2})-cos(\frac{-h}{2})}{h} \Rightarrow cos(\frac{h}{2})=cos(\frac{-h}{2}) \Rightarrow \boxed{\frac{cos(\frac{h}{2})-cos(-\frac{h}{2})}{h} \equiv 0}$

Just as $\frac{d}{dx}cos(x)|_{x=0}=-sin(0)=0$

This is NOT the case for the forward difference because the forward difference scheme is not symmetric about $x=0$.
